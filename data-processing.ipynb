```python
import csv
import pandas as pd

male_names = []
female_names = []
unique_names = []

# Example to access specific columns and handle headers
with open('Indian-Male-Names.csv', 'r', newline='') as csvfile:
    reader = csv.reader(csvfile)

    # Read the header row (optional)
    header = next(reader)
    print(f"Header: {header}")

    # Process data rows
    for row in reader:
        # Access values by index
        tempNames = row[0].split()
        male_names.append(tempNames[0] if tempNames else None)

    # unique_names = list(set(male_names))
    # print(len(unique_names))

with open('Indian-Female-Names.csv', 'r', newline='') as csvfile:
    reader = csv.reader(csvfile)

    # Read the header row (optional)
    header = next(reader)
    print(f"Header: {header}")

    # Process data rows
    for row in reader:
        # Access values by index
        tempNames = row[0].split()
        female_names.append(tempNames[0] if tempNames else None)

    # unique_names = list(set(female_names))
    # print(len(unique_names))

all_names = male_names + female_names
unique_names = list(set(all_names))
print(len(unique_names))


```

```python
def is_english_word(word):
    if not isinstance(word, str):   # skip None or numbers
        return False
    if not word:                    # empty string
        return False
    return all('a' <= ch.lower() <= 'z' for ch in word)

english_unique_names = [w for w in unique_names if is_english_word(w)]
```

```python
len(english_unique_names)
```

```python
with open("unique_indian_first_names.csv", "w", newline="") as f:
    writer = csv.writer(f)
    for n in english_unique_names:
        writer.writerow([n])
```
